```python
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .master("local[*]") \
    .appName('test') \
    .getOrCreate()
```

```python
from pyspark.sql import Row
from datetime import datetime
from pyspark.sql.functions import date_format
from pyspark.sql.functions import to_date
```

```python
from pyspark.sql import Row

data = [
    Row(order_number=1, customer_number=1),
    Row(order_number=2, customer_number=2),
    Row(order_number=3, customer_number=3),
    Row(order_number=4, customer_number=3),
]

df = spark.createDataFrame(data)

df.show()
print(df.dtypes)
```

```python
df.createOrReplaceTempView("Orders")
```

```python
result = spark.sql(
    """
    select customer_number,rank()over(order by count(order_number) desc )as rk
    from Orders
    group by customer_number
    """
)
result.show()
```

```python
result = spark.sql(
    """
    select customer_number from (select customer_number,rank()over(order by count(order_number) desc )as rk
    from Orders
    group by customer_number) as result
    where rk = 1
    """
)
result.show()
```
